# Netflix Stock Analysis Project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from datetime import datetime
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Subhajeet-Das/Netfilix-Stock-Price-prediction/main/NFLX.csv")
```

```python
data
```

```python
data.head()
```

# index replace with date

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
data['Date']= pd.to_datetime(data['Date'])
data = data.set_index('Date')
data.head()
```

# Volume of stock traded

```python
sns.lineplot(x= data.index, y = data['Volume'],label = 'Volume')
plt.title('Volume of stock versus time')
```

# Netflix stock Price- High, Open,Close

```python
data.plot(y = ['High','Close','Open'],title='Netflix Stock Price')
```

# Netflix Stock Price- Day,Months,Year wise

```python
fig,(ax1,ax2,ax3) = plt.subplots(3,figsize=(15,10))
data.groupby(data.index.day).mean().plot(y ='Volume',ax=ax1, xlabel='Day')
data.groupby(data.index.month).mean().plot(y ='Volume',ax=ax2, xlabel='Months')
data.groupby(data.index.year).mean().plot(y ='Volume',ax=ax3, xlabel='Year')
```

# Top-5 Dates with Highest Stock Price

```python
a = data.sort_values(by = 'High', ascending =False).head(5)
a['High']
```

```python
b = data.sort_values(by = 'Low', ascending =True).head(5)
b['Low']
```

```python
fig,axes =plt.subplots(nrows=1,ncols=2, sharex =True,figsize=(12,5))
fig.suptitle('High & Low values stock per period of time ')
sns.lineplot(ax=axes[0],y = data['High'],x = data.index,color='Green')
sns.lineplot(ax=axes[1],y = data['Low'],x = data.index,color='red')
```
